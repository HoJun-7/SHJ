```python
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

```python
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/suninno/healthcare_bigdata_playbook_kr/refs/heads/main/ch08/OTg6QzA_activities.csv"
df = pd.read_csv(url, delimiter = ",")
```

```python
df
```

```python
df.shape
```

```python
X = df.X.to_numpy()
Y = df.Y.to_numpy()
```

```python
X
```

```python
Y
```

```python
plt.style.use('bmh')
plt.figure(facecolor = 'white')

plt.scatter(X, Y)
plt.show()
```

```python
plt.figure(facecolor = 'white')
plt.hist2d(X, Y, bins = (25, 25), cmap = plt.cm.Reds)
cb = plt.colorbar()
cb.set_label('frequency of appearance, bins=25')
```

```python
plt.figure(facecolor='white')
plt.hist2d(X, Y, bins = (40, 40), cmap = plt.cm.Greens)
cb = plt.colorbar()
cb.set_label('frequency of appearance, bins=40')
```

```python
plt.figure(facecolor = 'white')
plt.hexbin(X, Y, gridsize = 20, cmap = plt.cm.Blues)
cb = plt.colorbar()
cb.set_label('frequency of appearance, gridsize=20')
```

```python
x1 = X[:16570] # 0 ~ 16569 xê°’
x2 = X[16570:16570*2] # 16570 ~ 33139 xê°’
x3 = X[16570*2:16570*3] # 33140 ~ 49709 xê°’
x4 = X[16570*3:] # 49710 ~ 66279 xê°’

y1 = Y[:16570] # 0 ~ 16569 yê°’
y2 = Y[16570:16570*2] # 16570 ~ 33139 yê°’
y3 = Y[16570*2:16570*3] # 33140 ~ 49709 yê°’
y4 = Y[16570*3:] # 49710 ~ 66279 yê°’
```

```python
plt.figure(facecolor='white')

# 1-Q(1ë“±ë¶„)
plt.subplot(2, 2, 1)
plt.hist2d(x1, y1, bins = 20, cmap = plt.cm.inferno)
plt.title('1-Q')
plt.colorbar()

# 2-Q(2ë“±ë¶„)
plt.subplot(2, 2, 2)
plt.hist2d(x2, y2, bins = 20, cmap = plt.cm.plasma)
plt.title('2-Q')
plt.colorbar()

# 3-Q(3ë“±ë¶„)
plt.subplot(2, 2, 3)
plt.hist2d(x3, y3, bins = 20, cmap = plt.cm.magma)
plt.title('3-Q')
plt.colorbar()

# 4-Q(4ë“±ë¶„)
plt.subplot(2, 2, 4)
plt.hist2d(x4, y4, bins = 20, cmap = plt.cm.cividis)
plt.title('4-Q')
plt.colorbar()

plt.show()
```
