**문제 9**

두 개의 1차원 어레이가 다음과 같이 주어졌다.

````python
A = np.ones(3)*1
B = np.ones(3)*2
````

아래 코드의 실행 결과를 설명하라.

````python
(A+B)*(-A/2)
````

`코드설명`: A 배열의 경우 3개의 요소가 전부 1로 되있는 배열을 생성 후 전부에 1을 곱해주고 B 배열의 경우 3개의 요소가 전부 1로 되있는 배열을 생성 후 전부에 2를 곱해준다. 그 뒤에 (A+B)*(-A/2)를 계산하게 되면 ([3, 3, 3]) * ([-0.5, -0.5, -0.5]) = ([-1.5, -1.5, -1.5])가 된다.

````python
np.add(A,B,out=B)
````

`코드설명`: 배열 A = ([1, 1, 1])와 배열 B = ([2, 2, 2])를 더한 값 ([3, 3, 3])을 배열 B에 값을 넣는다라는 개념이다.

````python
np.divide(A,2,out=A)
````

`코드설명`: 배열 A=([1,1,1])를 2로 나누기한 ([0.5, 0.5, 0.5])를 배열 A에 값을 넣는다라는 개념이다.

````python
np.negative(A,out=A)
````

`코드설명`: 배열 A=([1, 1, 1])를 부호 반전을 시켜 ([-1, -1, -1])로 만들고 배열 A에 값을 넣는다라는 개념이다.

````python
np.multiply(A,B,out=A)
````

`코드설명`: 배열 A = ([1, 1, 1])와 배열 B = ([2, 2, 2])를 곱한 값 ([2, 2, 2])를 배열 A에 값을 넣는다라는 개념이다.

**문제 7**

무작위 수로 구성된 (5, 5) 모양의 2차원 어레이가 다음과 같다.

````python
# 지정한 범위로 값을 조정하는 것 : 정규화, -1 ~ 1, 0.5
# 대원칙 : 인공지능은 학습전에 데이터가 정규화/표준화가 되어있어야 한다.
Z = np.random.random((5, 5))
````

변수 `Z`가 가리키는 2차원 어레이에 대해 min-max 크기조정(정규화)을 적용하라.

min-max 크기조정(정규화)은 항목의 모든 값을 아래 식을 통해 0과 1 사이의 부동소수점으로 변환한다. $max$와 $min$은 각각 최대 항목과 최소 항목을 가리킨다.

$$
\frac{x-min}{max - min}
$$

````python
# None을 적절한 표현식으로 대체하라.
Z = (Z - Z.min()) / (Z.max() - Z.min())
Z
````

아래 코드는 수정하지 마세요!!!

````python
assert (Z.max() == 1 and Z.min() == 0)
````
