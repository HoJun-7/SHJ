```python
# 코드를 작성하여라.
df1['K'] = df1['B'] * 3
df1
```

(5) `df1`에 `R`열을 추가하여라.
이때, `R`열의 각 항목은 균등분포에 따라 `[0, 1)` 구간에서 선택된 임의의 수로 구성된다.

(6) `df1`에서 `C`열에 인덱스가 1, 3, 5인 자료의 값을 6으로 변경하여라.

```python
# 코드를 작성하여라.
```

함수 호출에 사용된 키워드 인자의 의미는 다음과 같다.
